```python
# Read CSV file
df = pd.read_csv('data/FoodData_Central_foundation_food_csv_2024-10-31/food.csv',low_memory=False)

# Filter for sample_food data type only
df = df[df['data_type'] == 'sample_food']

# Remove duplicates in the 'description' column
df = df.drop_duplicates(subset=['description'])

# New 'id' column based on 'fdc_id', converted to string
df['id'] = df['fdc_id'].astype(str)

# New food_category column matching the id based on another table 'cat_df'
df['food_category'] = df['food_category_id'].map(cat_df.set_index('id')['description'])

# External url for each row
df['external_url'] = "https://cdn.britannica.com/36/123536-050-95CB0C6E/Variety-fruits-vegetables.jpg"

# Get 1st word of description column for item title
def get_title(row):
    for i in range(len(row['description'])):
        if (row['description'][i] == ","):
            return row['description'][:i]

df['title'] = df.apply(get_title, axis=1)

# Drop unnecessary column
df = df.drop(columns=['publication_date', 'data_type', 'food_category_id', 'fdc_id'])

# Display dataframe
df
```
